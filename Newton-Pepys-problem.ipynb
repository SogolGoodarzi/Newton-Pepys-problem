Evaluating the convergence value of tossing cubes:

```python
#With the use of equation:
import numpy as np
from math import factorial as fact
import matplotlib.pyplot as plt
pr=[]
n=np.array(np.arange(1,200))
for i in range(0,len(n)):
    x_=np.array(np.arange(0,n[i]))
    s=0
    for x in x_:
        s+=fact(6*n[i])/(fact(6*n[i]-x)*fact(x))*(1/6)**x*(5/6)**(6*n[i]-x)
    pr.append(1-s)
plt.plot(n,pr)
plt.ylim(0.45,0.7)
plt.xlabel("n")
plt.ylabel("pr")
plt.title("Probability of having at least n 6's in 6n tosses")
plt.plot(n,[0.5 for k in range(0,len(n))])
print("The value of convergence:",(sum(pr)/len(pr)))
```

```python
#With the use of simulation:
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
pr=[]
n=np.array(np.arange(1,200))
for i in range(0,len(n)):
    count=0
    for k in range(0,100):
        p_=[]
        c=0
        for j in range(0,6*n[i]):
            r=random.randint(1,7)  #choosing a random number for a toss
            p_.append(r)
        for a in range(0,len(p_)):
            if p_[a]==6:c+=1       #checking the number of 6's
        if c>=n[i]:count+=1        #checking the correct number of 6's we want
    pr.append(count/(100))
plt.plot(n,pr)
plt.ylim(0.2,0.8)
plt.xlabel("n")
plt.ylabel("pr")
plt.title("Probability of having at least n 6's in 6n tosses")
print("The value of convergence:",(sum(pr)/len(pr)))
```
